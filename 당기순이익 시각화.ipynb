```python
folder_path = './csv'
keyword = '손익계산서'

file_list = os.listdir(folder_path)
is_files = [file for file in file_list if keyword in file]

is_files
```

```python
d={}
for filename in is_files:
    df=pd.read_csv('./csv/{}'.format(filename),encoding='euc-kr')
    df_boolean = df=='당기순이익(손실)'
    x=df_boolean.sum(axis=1).values.argmax()
    y=df_boolean.sum(axis=0).values.argmax()
    ni=df.iloc[x,y+1]
    d[filename[11:18]]= int(ni)
```

```python
d
```

```python
sorted(d)
```

```python
d_new={}
summation=0
for i, period in zip(range(30), sorted(d)[1:]):
    remain=i%4
    if remain ==3:
        print(i, period, d[period]-summation)
        d_new[period]=d[period]-summation
        summation=0
       
    else:
        summation=summation+d[period]
        print(i, period, d[period])
     
        d_new[period]=d[period]
```

```python
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))  # 사이즈 조정은 가장 처음에 해야함
plt.plot(d_new.values())
plt.title("삼성전자 당기순이익 추이", fontsize=30)
plt.xlabel("기간")
plt.ylabel("당기순이익")
plt.grid()
plt.xticks(np.arange(30),d_new.keys(),rotation=90)
```
