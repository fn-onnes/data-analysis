```python
##여러 페이지 합치기## for문 + concat

url="https://finance.naver.com/sise/sise_market_sum.naver?&page="
pages=range(1,5)

df= pd.DataFrame([])
for i in pages:
    df1=pd.read_html(url+str(i), encoding='euc-kr')[1]
    df1["page"]=i
    df=pd.concat([df, df1])

df
```

```python
df = df.drop(df[df['N'].isna()].index)  ## N열의 NAN 값인 것을 불리안으로 받고, 그것이 TRUE가 되는 인덱스를 드랍해줘
```
