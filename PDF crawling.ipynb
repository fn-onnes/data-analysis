```python
import fitz  # PyMuPDF 라이브러리

def extract_text(path):      
    doc = fitz.open(path)  #pdf파일 열기
    text = ""

    for page_num in range(doc.page_count):   #페이지 수만큼 루프
        page = doc[page_num]                 #페이지들을 가져옴
        text += page.get_text()              #페이지마다 텍스트 따옴

    return text

# 예시: PDF 파일 경로 .   #여기서 경로를 이제 자동으로 따올 수 있어야함.
path = "신한은행.pdf"

# PDF에서 텍스트 추출
pdf_text= extract_text(path)

# 추출된 텍스트 출력
print(pdf_text)   #프린트로 출력하면 글자 깔끔하게 보임.
```

```python
import pdfplumber

def extract_pages(path, start_page):
    extracted_text = ""
    with pdfplumber.open(path) as pdf:
        for page_num in range(start_page+2,start_page+3):
            page = pdf.pages[page_num] #205p
            text = page.extract_text() #205페이지 내용 추출을 text에 담음
            extracted_text += text     #이후 연장되는 추가 가능 
    return extracted_text

# PDF 경로와 추출할 페이지 범위 설정
path = "신한은행.pdf"
start_page = 202
#end_page = 203  예시로 다음 페이지까지 포함

# 함수 호출
result = extract_pages(path, start_page)

# 결과 출력
print(result)
```
