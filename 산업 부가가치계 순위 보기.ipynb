```python

#한국은행 통계 분석 자료집을 공부하던 도중, '산업연관표'의 생산자 가격이 일종의 손익계산서 비용 역할을 한다는 점을 배우고,
#통계표를 보던 중 산업별 부가가치 순위를 보고 싶어서 해보았다.
```

```python
#불러오기
import pandas as pd
df=pd.read_excel("생산자가격.xlsx", skiprows=5)
```

```python
#결측치 제거
df_filled=df.fillna(0)
```

```python
#정렬위해 전치
df1=df_filled.T
```

```python
# 인위적으로 설정된 행을 열로 설정해주기 (drop위해)
df1.rename(columns=df1.iloc[0],inplace=True)
```

```python
#drop 
df=df1.drop(df1.index[0])
```

```python
#한번 더 반복
df.rename(columns=df.iloc[0],inplace=True)
```

```python
df1=df.drop(df.index[0])
```

```python
#정렬
df1.sort_values(by=['부가가치계'], ascending=False)
```
